### cleaning content

```python
import re
```

```python
df['Content'].iloc[1]
```

```python
def clean_content(text):
    pattern = '[a-zA-Z0-9]+'
    matched = re.findall(pattern, str(text))
    return ' '.join(matched).lower()
```

```python
clean_content(df['Content'].iloc[1])
```

```python
df['Clean_content'] = df['Content'].apply(clean_content)
```

### Title encoding

```python
import pickle
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))
```

```python
tfidf = vectorizer.fit_transform(df['Title_new'])
```

```python
with open('models/tfidf_values.pkl', 'wb') as file:
    pickle.dump(tfidf, file)
```

```python
with open('models/vectorize_search.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)
```

```python
title = input()
title = title.lower()
query_vec = vectorizer.transform([title])
similarity = cosine_similarity(query_vec, tfidf).flatten()
search_index = similarity.argsort()[-1:][::-1]
df.iloc[search_index]
```

```python
def search_by_name(movie_name):
    title = movie_name.lower()
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    search_index = similarity.argsort()[-1:][::-1]
    return df.iloc[search_index]
```

### Context encoding

```python
df1 = df.drop(['ipcratingstarrate'], axis=1)
```

```python
df1.head()
```

```python
df1.to_csv('data/tamil_movies_1990_2013_cleaned.csv')
```

```python
test_context = df1['Clean_content'].iloc[56]
test_context
```

```python
clean_context = df1['Clean_content'].tolist()
```

```python
embaddings = model.encode(clean_context)
```

```python
embaddings.shape
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
k = 6
knn_model = NearestNeighbors(n_neighbors=k, algorithm='auto')
knn_model.fit(embaddings)
```

```python
with open('models/knn_model.pkl', 'wb') as file:
    pickle.dump(knn_model, file)
```

```python
def recommend_movies(context):
    # Find k nearest neighbors
    embadded_context = model.encode(context)
    distances, indices = knn_model.kneighbors([embadded_context])
    
    # Return the indices of the nearest neighbors (movies)
    print(indices)
    return indices[0]
```

```python
idx = recommend_movies(test_context)
```
